This notebook is created to add headers to the GO_terms.tsv files and to the ones inside the networks folder. To accomplish this, a few lines in bash will be needed, it can also be added using python, but it is easier to add a line at the beginning of a file with bash than to read the whole file with python and modify it.

```python
%%bash

# Adding header GO file. Note we are modifying inplace
sed -i '1s/^/category\tGO\tgo_description\n/' GO_terms.tsv
head GO_terms.tsv
```

```python
%%bash

# We are gonna do the same for the networks

# 1. The id of this interaction
# 2. Transcription Factor TF
# 3. Target TG
# 4. Status of the interactions (is it known or is it new?)
# 5. Organism where that interaction was mapped
# 6. id of the interaction mapped in the organism where that interaction was mapped
# 7. Total organism where that interaction came from
# 8. Foward or Reverse string
# 9. Transcription Unit id of the target

for net in networks/*tsv; do
    
    sed -i '1s/^/id_net\ttf\ttg\tstatus\torg_reference\tid_net_reference\tcount_org_reference\tstring\ttu_unit\n/' "${net}"

done
```
